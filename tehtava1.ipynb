The values of the class column should be recoded as a or c (affected or control).

```python
df['class'].unique()
```

```python
df['class']= pd.Categorical(df['class'])
```

```python
df['class'] = df['class'].astype(str)
df['class'] = df['class'].replace({'ckd': 'a', 'ckd\t': 'a', 'notckd': 'c'})
df['class'] = df['class'].astype('category')
```

Rows with three or more missing values should be removed. Indicate the number of rows left in the modified data frame.

```python
df = df[df.isna().sum(axis=1) < 3]
df.info()
```

...and provide clear, readable histograms for each numerical column.

```python
def plot_data(df):
    plt.figure(figsize=(20, 10))
    for i, col in enumerate(df.select_dtypes(include=['int', 'float'])):
        plt.subplot(3, 5, i+1)
        df[col].plot(kind='hist', title=col)
        plt.xlabel(col)
    
    plt.subplots_adjust(hspace=0.4)
```

```python
plot_data(df1)
```

Most extreme values observed in the histograms fall within clinically plausible ranges and were therefore retained.
However, the potassium (pot) feature contains a value above 40, which is physiologically implausible and likely represents a data entry error.
This observation was treated as an outlier and removed from the dataset in next step.

```python
df1 = df1[df1['pot'] <= 10]

plot_data(df1)
```

```python
plot_data(df2)
```

In this dataset, no clearly implausible outliers are observed in the histograms.
The distributions appear compact and fall within clinically plausible ranges.
Therefore, no additional outlier removal was performed beyond the previously identified potassium values.
